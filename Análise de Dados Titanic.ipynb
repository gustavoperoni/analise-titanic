```python
# importar o pandas e matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar o arquivo base do titanic(para alteração)
titanic = pd.read_csv('train.csv')
```

```python
#Excluindo algumas colunas desnecessarias
titanic.drop(['PassengerId','SibSp','Parch','Ticket','Cabin','Embarked','Fare'],axis=1,inplace=True, )
```

```python
#Renomeando as colunas para português
titanic.rename(columns={'Survived':'Sobrevivente',
                       'Pclass':'Classe',
                       'Name':'Nome',
                       'Sex':'Sexo',
                        'Age':'Idade'},inplace=True)
```

```python
# Substituindo a "female" e "male" pelo masculino e feminino na coluna  "Sexo"
titanic.replace({'male':'masculino',
                'female':'feminino'}, inplace=True)
```

```python
# Vendo todos os valores valores únicos em cada coluna
titanic.nunique()
```

```python
# quantos valores vazios estão faltando
titanic.isnull().sum()
```

```python
# Vendo a descrição da coluna "Idade" para pegar a mediana ou a média para colocar nos valores faltantes. 
#OBS: usei a mediana pois a idade min é 0.42 e idade max é 80, ou seja, existem outliers que poderiam distorcer a média. 
titanic.Idade.describe()
```

```python
#Para ficar um preenchimento mais preciso irei utilizar a mediana do gênero e  classe
titanic.groupby(['Classe','Sexo'])['Idade'].median()
```

```python
# Usando o comando "transform" ('median') calculou a mediana da idade de cada grupo (Classe + Sexo) para cada linha e 
# fillna() substituiu apenas os valores NaN da coluna Idade por essas medianas de grupo.
titanic['Idade'] = titanic['Idade'].fillna(
    titanic.groupby(['Classe','Sexo'])['Idade'].transform('median')
)
```

```python
# Sobrevivência por gênero
titanic.groupby('Sexo')['Sobrevivente'].mean()
```

```python
# Sobrevivência por classe
titanic.groupby('Classe')['Sobrevivente'].mean()
```

```python
# Criando faixas de idade para facilitar a análise da sobrevivência por grupos etários
titanic['Faixa_Idade'] = pd.cut(
    titanic['Idade'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Criança', 'Adolescente', 'Jovem', 'Adulto', 'Idoso']
)
```

```python
# Calculando a taxa de sobrevivência por faixa etária
titanic.groupby('Faixa_Idade', observed=True)['Sobrevivente'].mean()
```

```python
# Sobrevivência por gênero
titanic.groupby('Sexo')['Sobrevivente'].mean().plot(kind='bar')

plt.title('Sobrevivência por Gênero')
plt.ylabel('Taxa de Sobrevivência')
plt.xlabel('Sexo')
```

```python
# Sobrevivência por classe
titanic.groupby('Classe')['Sobrevivente'].mean().plot(kind='bar')

plt.title('Sobrevivência por Classe')
plt.ylabel('Taxa de Sobrevivência')
plt.xlabel('Classe')

plt.show()
```

```python
# Sobrevivência por faixa etária
titanic.groupby('Faixa_Idade', observed=True)['Sobrevivente'].mean().plot(kind='bar')

plt.title('Sobrevivência por Faixa Etária')
plt.ylabel('Taxa de Sobrevivência')
plt.xlabel('Faixa Etária')

plt.show()
```
